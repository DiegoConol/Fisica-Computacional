```python
import os
import re
import matplotlib.pyplot as plt

archivos = []
patron = re.compile(r"lyapunov_([0-9]+\.[0-9])\.txt")

# Busca archivos que coincidan con el patrn
for nombre in os.listdir('.'):
    if patron.match(nombre):
        archivos.append(nombre)

archivos.sort()  # Opcional: ordena los archivos por nombre

for archivo in archivos:
    E = patron.match(archivo).group(1)
    y_vals = []
    with open(archivo) as f:
        for linea in f:
            if linea.strip():
                y = float(linea.split()[0])
                y_vals.append(y)
    plt.plot(y_vals, label=f"E={E}")

plt.xlabel("ndice de tiempo")
plt.ylabel("Lyapunov")
plt.title("Exponentes de Lyapunov")
plt.legend()
plt.tight_layout()
plt.show()
```
