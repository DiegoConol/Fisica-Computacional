## Informe Estabilidad

En este notebook se presentan los resultados de la estabilidad tanto para variaciones de energía como de posición.

### Programa 1: Estabilidad para variaciones de ángulos iniciales.

Aquí se variarán los ángulos iniciales y se pondrán los mapas de Poincaré correspondientes para diferencias pequeñas puesto que estamos viendo la estabilidad del sistema.

Para ello supondremos una diferencia angular de 0.05 radianes. 

Usaremos el programa 'estabilidad.c', que suelta solo la parte que mide los ángulos. Para cada caso mediremos los ángulos con la diferencia y sin esta.

#### Programa 1.1: Estabilidad para variación del ángulo $\theta$

En este apartado veremos lo que le pasa al mapa de poincaré si cambiamos un poco el ángulo $\theta$, que es el ángulo más cercano al punto fijo del péndulo doble.

El programa necesita 2 ficheros txt, siendo uno de ellos el aplicado la diferencia y otro sin ella.


```python
import numpy as np
import matplotlib.pyplot as plt

# Leer los datos de los archivos
datos_1 = np.loadtxt('angulos_1.1.txt')
datos_2 = np.loadtxt('angulos_1.2.txt')

# Graficar
plt.figure(figsize=(8, 5))
plt.plot(datos_1[:, 0], datos_1[:, 1], label='Ángulos iguales', color='blue', alpha=0.7)
plt.plot(datos_2[:, 0], datos_2[:, 1], label='Ángulos con diferencia', color='red', alpha=0.7)
plt.xlabel(r'$\theta$')
plt.ylabel(r'$\phi$')
plt.title(r'Mapa de Poincaré para $\theta$ = $\phi$ = $\pi$/4 y diferencia de 0.05 radianes en $\theta$ y E=10')
plt.legend()
plt.grid(True)
plt.savefig("Estabilidad_angular_1.png")
plt.show()
```

En este ploteo podemos observar que no tenemos mucha estabilidad. En los bordes hay superposición perfecta de los péndulos, pero en la parte central no es tan así. Vemos como la curva roja presenta separaciones más equidistantes, mientras que la azul parece que los ángulos que puede tomar son más fijos, es decir, la curva se junta más en unas zonas y no está en otras.



#### Programa 1.2: Estabilidad para variación del ángulo $\phi$

En este apartado veremos lo que le pasa al mapa de poincaré si cambiamos un poco el ángulo $\phi$, que es el ángulo más alejado al punto fijo del péndulo doble.

El programa necesita 2 ficheros txt, siendo uno de ellos el aplicado la diferencia y otro sin ella.

```python
import numpy as np
import matplotlib.pyplot as plt

# Leer los datos de los archivos
datos_1 = np.loadtxt('angulos_2.1.txt')
datos_2 = np.loadtxt('angulos_2.2.txt')

# Graficar
plt.figure(figsize=(8, 5))
plt.plot(datos_1[:, 0], datos_1[:, 1], label='Ángulos iguales', color='blue', alpha=0.7)
plt.plot(datos_2[:, 0], datos_2[:, 1], label='Ángulos con diferencia', color='red', alpha=0.7)
plt.xlabel(r'$\theta$')
plt.ylabel(r'$\phi$')
plt.title(r'Mapa de Poincaré para $\theta$ = $\phi$ = $\pi$/4 y diferencia de 0.05 radianes en $\phi$ y E=10')
plt.legend()
plt.grid(True)
plt.savefig("Estabilidad_angular_2.png")
plt.show()
```

Como iba diciendo en el anterior ploteo, hay zonas donde una curva está más aglotinada que la otra. En el caso anterior era la azul la que mostraba esta "cuadricula" interior, pero ahora es la roja la que presenta una mayor. Por ello podemos concluir que no tenemos una estabilidad bien definida, al menos para esta energía de E =10.0


### Programa 2: Estabilidad energética.

El fundamento es el mismo que el anterior. Sin embargo es ahora el turno de la energía. Para ello aplicaremos una diferencia de 0.5 unidades. Esto es, que el segundo péndulo tiene 0.5 unidades más que el primero, de 10 unidades.

```python
import numpy as np
import matplotlib.pyplot as plt

# Leer los datos de los archivos
datos_1 = np.loadtxt('angulos_3.1.txt')
datos_2 = np.loadtxt('angulos_3.2.txt')

# Graficar
plt.figure(figsize=(8, 5))
plt.plot(datos_1[:, 0], datos_1[:, 1], label='Energía E=10.0', color='blue', alpha=0.7)
plt.plot(datos_2[:, 0], datos_2[:, 1], label='Energía E=11.0', color='red', alpha=0.7)
plt.xlabel(r'$\theta$')
plt.ylabel(r'$\phi$')
plt.title(r'Mapa de Poincaré para $\theta$ = $\phi$ = $\pi$/4 y diferencia de $\Delta$E=0.5')
plt.legend()
plt.grid(True)
plt.savefig("Estabilidad_energetica.png")
plt.show()
```

De nuevo, no podemos concluir que exista una estabilidad bien definida. Si bien los péndulos pueden moverse parecido, hay regiones de ángulos que no toman, como en el caso de la separación angular.

Sin embargo, como los bordes casi son identicos, podemos decir que está bien acotado y por tanto sí existe la estabilidad de no estar en un régimen caótico.

### Programa 3: Y si combinamos energía y ángulos

El planteamiento es el mismo que en los casos anteriores, con una energía inicial de E=10, ángulos iniciales iguales a pi/4 y una diferencia de energía de 0.5 unidades y angular de 0.05 radianes, observaremos el mapa de poincaré superpuesto de ambos péndulos dobles.

```python
import numpy as np
import matplotlib.pyplot as plt

# Leer los datos de los archivos
datos_1 = np.loadtxt('angulos_4.1.txt')
datos_2 = np.loadtxt('angulos_4.2.txt')

# Graficar
plt.figure(figsize=(8, 5))
plt.plot(datos_1[:, 0], datos_1[:, 1], label='Energía E=10.0', color='blue', alpha=0.7)
plt.plot(datos_2[:, 0], datos_2[:, 1], label='Energía E=11.0', color='red', alpha=0.7)
plt.xlabel(r'$\theta$')
plt.ylabel(r'$\phi$')
plt.title(r'Mapa de Poincaré para $\theta$ = $\phi$ = $\pi$/4 diferencia de $\Delta$E=0.5')
plt.legend()
plt.grid(True)
plt.savefig("Estabilidad_energetica.png")
plt.show()
```

No podemos observar mucho, puesto que el cambio de energía domina, no se puede apreciar con claridad la diferencia de ángulos, que si bien es pequeña sí se ha podido ver el caso aislado.
